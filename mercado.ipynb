```python
import pandas as pd
```

```python
df_mercado = pd.read_csv("./data/sales_curva_a.csv", delimiter=',', encoding='UTF-8',low_memory=False)
df_mercado
```

```python
itens = df_mercado['item_nbr'].unique().tolist()
top_itens = df_mercado['item_nbr'].value_counts().head(10).index
total_itens = len(itens)
print("Lista de itens: ", itens)
print("Top 10 itens: ", top_itens)
print("Total de itens: ", total_itens)
```

```python
df_mercado_top = df_mercado[df_mercado['item_nbr'].isin([top_itens[0]])]
df_mercado_top
```
